# **Regression Modeling( Continous Data)**

# Linear regression is a predictive modeling technique. 
# It is used whenever there is a **linear relation** between the dependent and the independent variables.

##                  Y = b0 + b1* x 

# It is used in estimating exactly how much of y will change, when x changes a certain amount.

# **Example**

## **Import Data Manipulation and Data Visulaization Libraries**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**1. Load the Supermarket  Dataset**

```
data=pd.read_csv("/content/Supermarket.csv")
print(data)
```

**2. Understand the Data**

```
data.info()
```

```
data.describe()
```

3. **Define X and Y values**

```
# Collecting X and Y
X = data['Sales of food Items ( X1)'].values
Y = data['Netprofit(In crores)'].values
```

```
print(X)
```

```
print(Y)
```

**4. Calulate Mean of X and Y**

```
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```
print("Mean_X",mean_x)
print("Mean_Y",mean_y)
```

```
n = len(X)
print(n)
```

**5.Use the formula to calculate bo and b1**

```

numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
    b1 = numer / denom
    b0 = mean_y - (b1 * mean_x)
# Print coefficients and Intercept
print("Co-efficient",b1)
print("Intercept",b0)    

 
```

**6. Plotting Values and Regression Line**

```
max_x = np.max(X)
min_x = np.min(X) 
```

**7. Calculating line values x and y**

```

x = np.linspace(min_x, max_x, 10)
y = b0 + b1 * x 
print(x.shape)
print(y.shape)
```

**8. Plot Regression Line and Scatter points**

```
# Ploting Line
plt.plot(x, y, color='#52b920', label='Regression Line')
# Ploting Scatter Points
plt.scatter(X, Y, c='#ef4423', label='Scatter Plot')
plt.xlabel('Sales of food Items ( X1)')
plt.ylabel('Netprofit(In crores)')
plt.legend()
plt.show()
```

# How does residual help in finding the best fit line?

# To find out the best fit line, we have something called residual sum of squares (RSS). In RSS, we take the square of residuals and sum them up.

# **Root mean squared error (RMSE) **
# Root mean squared error (RMSE) is the square root of the mean of the square of all of the error. 

# It is otherwise called as excellent general-purpose error metric for numerical predictions.

# **The term is always between 0 and 1, since r is between -1 and 1**

```
#ss_t is the total sum of squares and ss_r is the total sum of squares of residuals(relate them to the formula).
ss_t = 0
ss_r = 0
for i in range(n):
    y_pred = b0 + b1 * X[i]
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_pred) ** 2
    r2 = 1 - (ss_r/ss_t)
print(r2)
```
